```python
import folium 

m = folium.Map(location=[43.6047, 1.442], zoom_start=12)
```

```python
import csv 

with open('defibrillateurs-toulouse.csv', newline='', encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        next(csv_reader)
        for row in csv_reader:
            geoloc = row[0].split(',')
            folium.Marker(location=[float(geoloc[0]), float(geoloc[1])], popup=f'{row[3]}', icon=folium.Icon(color='red', icon='info-sign')).add_to(m)
```
